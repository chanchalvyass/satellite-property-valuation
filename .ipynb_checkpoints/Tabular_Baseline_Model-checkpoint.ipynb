```python
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.efficientnet import preprocess_input
```

```python
cnn_model = EfficientNetB0(
    weights="imagenet",
    include_top=False,
    pooling="avg",       
    input_shape=(224, 224, 3)
)
```

```python
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(224, 224))
    img = img_to_array(img)
    img = preprocess_input(img)
    return img
```

```python
IMAGE_DIR = "Satellite Imagess"

available_ids = {
    int(f.replace(".png", ""))
    for f in os.listdir(IMAGE_DIR)
    if f.endswith(".png") and f.replace(".png", "").isdigit()
}

df_with_images = df[df["id"].isin(available_ids)].copy()
print("Rows with images:", len(df_with_images))
```

```python
from tqdm import tqdm
BATCH_SIZE = 32   # safe for CPU/GPU; increase if you have GPU

image_features = []
valid_ids = []

batch_imgs = []
batch_ids = []

for image_id in tqdm(df_with_images["id"], desc="Extracting EfficientNet features"):
    image_path = f"{IMAGE_DIR}/{image_id}.png"
    if not os.path.exists(image_path):
        continue

    img = preprocess_image(image_path)
    batch_imgs.append(img)
    batch_ids.append(image_id)

    if len(batch_imgs) == BATCH_SIZE:
        batch_imgs = np.array(batch_imgs)
        feats = cnn_model.predict(batch_imgs, verbose=0)

        image_features.extend(feats)
        valid_ids.extend(batch_ids)

        batch_imgs, batch_ids = [], []

# process remaining images
if batch_imgs:
    batch_imgs = np.array(batch_imgs)
    feats = cnn_model.predict(batch_imgs, verbose=0)
    image_features.extend(feats)
    valid_ids.extend(batch_ids)
```

```python
image_features = np.array(image_features)
valid_ids = np.array(valid_ids)

print(image_features.shape)   # (16110, 2048)
print(len(valid_ids))         # 16110
```

```python
image_feature_df = pd.DataFrame(
    image_features,
    index=valid_ids
)
image_feature_df.index.name = "id"
```

```python
image_features = np.array(image_features)
print(image_features.shape)
```

```python
multimodal_df = df.join(image_feature_df, how="inner")


```

```python
multimodal_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=56, random_state=42)

image_features_pca = pca.fit_transform(image_feature_df.values)
```

```python
print(image_features_pca.shape)
print(f"Explained variance retained: {pca.explained_variance_ratio_.sum():.2%}")

```

```python
image_feature_pca_df = pd.DataFrame(
    image_features_pca,
    index=image_feature_df.index
)

image_feature_pca_df.index.name = "id"
```

```python
multimodal_df_pca = df.join(image_feature_pca_df, how="inner")
```

```python
import numpy as np

y = np.log(multimodal_df_pca["price"])
x = multimodal_df_pca.drop(columns=["price"])
x = x.drop(columns=["date"])

```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=400,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = xgb_model.predict(x_test)

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Multimodal (PCA) R2:", r2)
print("Multimodal (PCA) RMSE:", rmse)
```
